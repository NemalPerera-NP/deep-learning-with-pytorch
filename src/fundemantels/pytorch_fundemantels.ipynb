### Creating Tensors

PyTorch tensors are created using 'torch.Tensor()' 
A torch.Tensor is a multi-dimensional matrix containing elements of a single data type.

```python
# Scalar
scalar = torch.tensor(7)
scalar
```

```python
scalar.ndim #this gives the number of dimensions
```

```python
scalar.item() #Get the tensor back as Python int
```

```python
#Vector
vector = torch.tensor([7,7])
vector
```

```python
vector.ndim
# Dimension is the number of []
```

```python
# Random Tensers
#Create a random tensor of size
random_tensor = torch.rand(3,4)
random_tensor
```

```python
random_tensor.ndim
```

```python
random_tensor = torch.rand(2,3,3,4)
random_tensor
```

```python
random_tensor.ndim
```

```python
#Create a random tensor with similar shape to an image tensor
random_image_size_tensor = torch.rand(size=(224,224,3)) # height width color chanels
random_image_size_tensor.shape,random_image_size_tensor.ndim
```

```python
# tensor of all zeros
zeros = torch.zeros(size=(3,4))
zeros
```

```python
# tensor of all ones (1)
zeros = torch.ones(size=(3,4))
zeros
```

### creating a range of tensors and tensors-like

```python
torch.arange(0,10)
```

```python
with_gaps = torch.arange(0,1000,25) #(start,end,step)
with_gaps
```

```python
all_zeros = torch.zeros_like(input=with_gaps)
all_zeros
```

```python
# Float 32 tensor
float_32_tensor = torch.tensor([3.0,6.0,9.0],
                              dtype = None, # what datatype is the tensor
                              device = None, # what device is your tensor on
                              requires_grad = False) # whether or not track gradients with this tensor operations
float_32_tensor 
```
